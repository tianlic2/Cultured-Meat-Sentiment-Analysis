```python
from openai import OpenAI
import pandas as pd

client = OpenAI(
    api_key="XXX", #updated on 03132024
)

def answer(prompt):
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature = 0
    )
    return response.choices[0].message.content.strip()
```

The tweets were sent to GPT-4o one by one using the following prompt:

```python
def feed(i,df):
    text = ('''Here's a codebook for sentiment analysis of cultured-meat related tweets:'''
    + codebook +
    "\n" + df['Tweet'][i] + 'please only return the number associated with the sentiment')
    return text
```
